**Zadanie 3**

Zdefiniowanie oraz wyrysowanie funkcji $f(x) = x^2 + 5$ dla:

1. $-1 < x < 1$,

2. $-6 < x < 6$,

3. $0 < x < 5$.

```python
#deklaracja funkcji
def function(a, b, title, xlabel, ylabel, legend=''):
    # Parametry:
    # a - dolna wartość przedziału
    # b - górna wartość przedziału
    # title - tytuł wykresu
    # xlabel - etykieta osi X
    # ylabel - etykieta osi Y
    # legend - legenda
    
    # wyliczenie funkcji
    x = np.arange(a, b, 0.01)
    y = x**2 + 5
    
    # wykres
    fig = plt.figure(figsize=(4, 4))
    plt.plot(x, y)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend([legend])
    plt.grid()
    plt.show()
```

```python
function(-1, 1, 'Wykres funkcji f(x)=x^2+5', 'x', 'f(x)=y', '-1<x<1')
function(-6, 6, 'Wykres funkcji f(x)=x^2+5', 'x', 'f(x)=y', '-6<x<6')
function(0, 5, 'Wykres funkcji f(x)=x^2+5', 'x', 'f(x)=y', '0<x<5')
```
